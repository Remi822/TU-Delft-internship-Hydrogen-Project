# Section 1 : Heat equation in 1D with linear BC and explicit solver 

The goal here is to solve the heat equation (see below) in one dimension with linear boundary conditions on a segment domain **[0,1]** : 
$$
\frac{\partial u}{\partial t} = D \frac{\partial^2 u}{\partial x^2} + f(x,t)
$$

## Parameters choice 
**Thermal diffusivity : D = 1**  
In order to have a trivial solution to this problem, we choose the following **source term : f(x,t) = x**  
**Initial condition : u(x,0) = 0**  

**Linear BC**  
**- Left boundary**  
We choose a strong Dirichlet condition on the left side (x=0) : **u(0,t) = 0**  
**- Right boundary**  
On the right side (x=1), we choose a time-dependent linear boundary condiiton : **u(1,t) = t**. This condition respects the initial condition u(0,x) = 0




## Section 1.1 : Import packages 


```julia
using Ferrite
using SparseArrays
using LinearAlgebra         # provides mul! 
using UnPack                # added for time integration using DifferentialEquations.jl 
using OrdinaryDiffEq        # added for time integration using DifferentialEquations.jl
using DifferentialEquations # added for time integration using DifferentialEquations.jl   
using StaticArrays    

using WriteVTK
using Plots                 # provides contour plot of cell averages 
using NLSolvers
using ForwardDiff
using Printf   


```

## Section 1.2 : Definition of the mesh, the boundary conditions, the initial condition, the quadrature etc.

```julia
nels = (40,)
left = Vec((0.,))
right = Vec((1.,))
grid = generate_grid(Line, nels, left, right)

ip = Lagrange{Ferrite.RefLine, 1}()
qr = QuadratureRule{RefLine}(2)
cellvalues = CellValues(qr, ip);

dh = DofHandler(grid)
add!(dh, :u, ip)
close!(dh);

## Allocation of the useful matrix
K = allocate_matrix(dh)
M = allocate_matrix(dh)
# Condition initialize
u0 = zeros(ndofs(dh))  # condition initiale nulle

max_temp = 20
t_rise = 10
ch = ConstraintHandler(dh)
left_boundary = getfacetset(grid, "left")
right_boundary = getfacetset(grid, "right")

add!(ch, Dirichlet(:u, left_boundary, (x, t) -> 0.0))
add!(ch, Dirichlet(:u, right_boundary, (x, t) -> t))
close!(ch)

update!(ch, 0.0);
```

## Section 1.3 : Assembly

```julia
function assemble_K(K::SparseMatrixCSC,  cellvalues::CellValues, dh::DofHandler, t::Float64)
    n_basefuncs = getnbasefunctions(cellvalues)
    Ke = zeros(n_basefuncs, n_basefuncs)
    fe = zeros(n_basefuncs)
    f = zeros(ndofs(dh))
    # Allocate global force vector f
    # Create an assembler
    assembler_K = start_assemble(K, f) 
    
    for cell in CellIterator(dh)
        # Reinitialize cellvalues for this cell
        Ferrite.reinit!(cellvalues, cell)
        # Reset to 0
        fill!(Ke, 0)
        fill!(fe, 0)
        # Loop over quadrature points
        for q_point in 1:getnquadpoints(cellvalues)
            # Get the quadrature weight
            dΩ = getdetJdV(cellvalues, q_point)
            x = spatial_coordinate(cellvalues, q_point, getcoordinates(cell))[1]
            f_val = x
            # Loop over test shape functions
            for i in 1:n_basefuncs
                v = shape_value(cellvalues, q_point, i)
                ∇v = shape_gradient(cellvalues, q_point, i)
                # Add contribution to fe
                fe[i] += f_val * v * dΩ
                # Loop over trial shape functions
                for j in 1:n_basefuncs
                    ∇u = shape_gradient(cellvalues, q_point, j)
                    # Add contribution to Ke
                    k_val = 1.0
                    Ke[i, j] += (∇v ⋅ ∇u) * dΩ * k_val
                end
            end
        end
        # Assemble Ke and fe into K and f
        assemble!(assembler_K, celldofs(cell), Ke, fe)
    end
    return K, f
end

function assemble_M(cellvalues::CellValues, M::SparseMatrixCSC, dh::DofHandler)
    # Allocate the element stiffness matrix and element force vector
    n_basefuncs = getnbasefunctions(cellvalues)
    Me = zeros(n_basefuncs, n_basefuncs)
    # Create an assembler
    assembler_M = start_assemble(M)
    # Loop over all cels
    for cell in CellIterator(dh)
        # Reinitialize cellvalues for this cell
        Ferrite.reinit!(cellvalues, cell)
        # Compute element contribution
        fill!(Me, 0)
        # Loop over quadrature points
        for q_point in 1:getnquadpoints(cellvalues)
            # Get the quadrature weight
            dΩ = getdetJdV(cellvalues, q_point)
            # Loop over test shape functions
            for i in 1:n_basefuncs
                v = shape_value(cellvalues, q_point, i)
                # Loop over trial shape functions
                for j in 1:n_basefuncs
                    u = shape_value(cellvalues, q_point, j)
                    Me[i, j] += v * u *  dΩ 
                end
            end
        end
        # Assemble Ke and fe into K and f
        assemble!(assembler_M, celldofs(cell), Me)
    end
    return M
end
```

## Section 1.4 : Solving with DifferentialEquations.jl

We use the Tsit5 alogrithm for the solivng which is a Runge-Kutta method of order 5

```julia

M = assemble_M(cellvalues, M, dh)

function heat_equation_ode!(du, u, p, t)
    K, M, dh, cellvalues, ch = p
    K, f = assemble_K(K , cellvalues, dh, t )
    update!(ch, t);
    apply!(K, f, ch)
    du .= M\ (f - K * u)
end

tspan = (0.0, 20.0)


params = (K, M, dh, cellvalues, ch)
prob = ODEProblem(heat_equation_ode!, u0, tspan, params)
sol = DifferentialEquations.solve(prob, Tsit5(),saveat=0.1);
```

## Section 1.5 : Results validation  

In order to check that our simulation is correct, we compare the results of our simulation with the analytical results.
The heat equation we are solving is : 
$$
\frac{\partial u}{\partial t} = D \frac{\partial^2 u}{\partial x^2} + x
$$

With this source term, the trivial solution is : $u(x,t) = x t$. We can check by adding the derivatives :  
$$\frac{\partial u(x,t)}{\partial t} = \frac{\partial [x t]}{\partial t} = x$$
$$\frac{\partial^2 u(x,t)}{\partial x^2} = \frac{\partial^2 [x t]}{\partial x^2} = 0$$  
So, we have  
$$\frac{\partial u(x,t)}{\partial t} + \frac{\partial^2 u(x,t)}{\partial x^2} = x = f(x,t)$$  
Finally, the equation is indeed satisfied.  

Then, we plot the analytical solution and the approximated solution on the same graph for five different timesteps and we notice that the results are satisfactory.






```julia

# Results visualisation ####################################################################################
node_coords = [x.x[1] for x in getnodes(grid)]
analytical_solution(x, t) = x*t
plt = plot(xlabel="x (m)", ylabel="Temperature u(x,t)", 
           title="Transient Heat Equation 1D", 
           legend=:bottomright, size=(800,600))


# Choisir 5 indices répartis uniformément dans le vecteur de temps
n_times = 5
indices = round.(Int, range(1, length(sol.t), length=n_times))
for i in indices
    t_val = sol.t[i]
    u_num = sol.u[i]
    plot!(plt, node_coords, u_num, label="Numerical t=$(round(t_val, digits=2))", lw=2)
    scatter!(plt, node_coords, analytical_solution.(node_coords, t_val), label="Analytical t=$(round(t_val, digits=2))", lw=2, ls=:dash)
end

display(plt)
```

## Section 1.6 : Computation of the error

To finalise the validation of the results, we now plot the error between the analytical solution and our approximate solution. We plot the error with two different method. 

**Norm $L^2$ error**  
First of all, we plot the total error over the domain for every timestep using the $L^2$ norm which is : 

$$
\| u_{\text{ana}} - u_{\text{approx}} \|_{L^2} \approx \sqrt{ \sum_{i=1}^N \left( u_{\text{ana}}(x_i) - u_{\text{approx}}(x_i) \right)^2 \Delta x_i }
$$

**Norm $L^\infty$ error**  
In a second phase, we plot the total error over the domain for every timestep using another norm; the $L^\infty$ norm, which is : 

$$
\| u_{\text{ana}} - u_{\text{approx}} \|_{L^\infty} \approx \max_{1 \leq i \leq N} \left| u_{\text{ana}}(x_i) - u_{\text{approx}}(x_i) \right|
$$

with **N** the number of points in our discretisation.



```julia
# Computation of the error ########################################################################################
#  Computing L² error at the t time
function compute_L2_error(u_num, t, dh, cellvalues)
    err_squared = 0.0
    for cell in CellIterator(dh)
        Ferrite.reinit!(cellvalues, cell)
        coords = getcoordinates(cell)
        dofs = celldofs(cell)
        u_cell = u_num[dofs]

        for q in 1:getnquadpoints(cellvalues)
            dΩ = getdetJdV(cellvalues, q)
            x = spatial_coordinate(cellvalues, q, coords)[1]
            u_exact = analytical_solution(x, t)
            u_approx = 0.0
            for i in 1:getnbasefunctions(cellvalues)
                φ = shape_value(cellvalues, q, i)
                u_approx += u_cell[i] * φ
            end
            err_squared += (u_exact - u_approx)^2 * dΩ
        end
    end
    return sqrt(err_squared)
end

# computation of the L² error during the time and visualisation
L2_errors = [compute_L2_error(u, t, dh, cellvalues) for (u, t) in zip(sol.u, sol.t)]
plt1 = plot(sol.t, L2_errors,
     xlabel="Time (s)", ylabel="Error L²",
     title=" L² error evolution during the time",
     lw=2, legend=false)

display(plt1)

# Computation of the infinite error 
# Analytical solution 
u_exact(x, t) = x * t
# Table for storing the  ‖·‖∞ error at every timestep
errors_inf = Float64[]
times = Float64[]

for (u_num, t) in zip(sol.u, sol.t)
    u_ana = u_exact.(node_coords, t)
    err = abs.(u_num .- u_ana)
    push!(errors_inf, maximum(err))
    push!(times, t)
end

# Visualisation 
plt2 = plot(times, errors_inf,
     xlabel="Time t",
     ylabel="Error ‖u - u_exact‖∞",
     title="Maximal error (∞ norm) during the time ",
     lw=2,
     legend=false)

display(plt2)
```

# Section 2 : Heat equation in 1D with linear BC and implicit solver

## Section 2.1 Numerical Implementation

Exact same code as in Section 1

```julia
nels = (40,)
left = Vec((0.,))
right = Vec((1.,))
grid = generate_grid(Line, nels, left, right)

ip = Lagrange{Ferrite.RefLine, 1}()
qr = QuadratureRule{RefLine}(2)
cellvalues = CellValues(qr, ip);

dh = DofHandler(grid)
add!(dh, :u, ip)
close!(dh);

## Allocation of the useful matrix
K = allocate_matrix(dh)
M = allocate_matrix(dh)
# Condition initialize
u0 = zeros(ndofs(dh))  # condition initiale nulle

max_temp = 20
t_rise = 10
ch = ConstraintHandler(dh)
left_boundary = getfacetset(grid, "left")
right_boundary = getfacetset(grid, "right")

add!(ch, Dirichlet(:u, left_boundary, (x, t) -> 0.0))
add!(ch, Dirichlet(:u, right_boundary, (x, t) -> t))
close!(ch)

update!(ch, 0.0);


function assemble_K(K::SparseMatrixCSC,  cellvalues::CellValues, dh::DofHandler)
    n_basefuncs = getnbasefunctions(cellvalues)
    Ke = zeros(n_basefuncs, n_basefuncs)
    fe = zeros(n_basefuncs)
    f = zeros(ndofs(dh))
    # Allocate global force vector f
    # Create an assembler
    assembler_K = start_assemble(K, f) 
    
    for cell in CellIterator(dh)
        # Reinitialize cellvalues for this cell
        Ferrite.reinit!(cellvalues, cell)
        # Reset to 0
        fill!(Ke, 0)
        fill!(fe, 0)
        # Loop over quadrature points
        for q_point in 1:getnquadpoints(cellvalues)
            # Get the quadrature weight
            dΩ = getdetJdV(cellvalues, q_point)
            x = spatial_coordinate(cellvalues, q_point, getcoordinates(cell))[1]
            f_val = x
            # Loop over test shape functions
            for i in 1:n_basefuncs
                v = shape_value(cellvalues, q_point, i)
                ∇v = shape_gradient(cellvalues, q_point, i)
                # Add contribution to fe
                fe[i] += f_val * v * dΩ
                # Loop over trial shape functions
                for j in 1:n_basefuncs
                    ∇u = shape_gradient(cellvalues, q_point, j)
                    # Add contribution to Ke
                    k_val = 1.0
                    Ke[i, j] += (∇v ⋅ ∇u) * dΩ * k_val
                end
            end
        end
        # Assemble Ke and fe into K and f
        assemble!(assembler_K, celldofs(cell), Ke, fe)
    end
    return K, f
end

function assemble_M(cellvalues::CellValues, M::SparseMatrixCSC, dh::DofHandler)
    # Allocate the element stiffness matrix and element force vector
    n_basefuncs = getnbasefunctions(cellvalues)
    Me = zeros(n_basefuncs, n_basefuncs)
    # Create an assembler
    assembler_M = start_assemble(M)
    # Loop over all cels
    for cell in CellIterator(dh)
        # Reinitialize cellvalues for this cell
        Ferrite.reinit!(cellvalues, cell)
        # Compute element contribution
        fill!(Me, 0)
        # Loop over quadrature points
        for q_point in 1:getnquadpoints(cellvalues)
            # Get the quadrature weight
            dΩ = getdetJdV(cellvalues, q_point)
            # Loop over test shape functions
            for i in 1:n_basefuncs
                v = shape_value(cellvalues, q_point, i)
                # Loop over trial shape functions
                for j in 1:n_basefuncs
                    u = shape_value(cellvalues, q_point, j)
                    Me[i, j] += v * u *  dΩ 
                end
            end
        end
        # Assemble Ke and fe into K and f
        assemble!(assembler_M, celldofs(cell), Me)
    end
    return M
end
```

## Section 2.2 : Solving with an implicit solver
Here, we solve the problem using the Rodas5 solver which is an implicit solver of order 5 and stable even with very thin mesh

```julia

## solving 
K, f = assemble_K(K, cellvalues, dh)
K_buffer = copy(K)
f_buffer = copy(f)
M = assemble_M(cellvalues, M, dh)

struct ProblemParameters 
    K::SparseMatrixCSC{Float64, Int}
    M::SparseMatrixCSC{Float64, Int} # Garder M_global ici, elle sera passée à mass_matrix
    f::Vector{Float64}
    dh::DofHandler
    cellvalues::CellValues
    ch::ConstraintHandler # Contient seulement les CL linéaires
    K_buffer::SparseMatrixCSC{Float64, Int}
    f_buffer::Vector{Float64}
end

function heat_equation_ode!(du, u, p::ProblemParameters, t)
    @unpack K, M, f, dh, cellvalues, ch, K_buffer, f_buffer = p 
    copy!(K_buffer, K)
    copy!(f_buffer, f)
    update!(ch, t)
    apply!(K_buffer, f_buffer, ch)
    # second term construction 
    mul!(du, K_buffer, u)       # du = K_buffer * u
    du .*= -1.0
    du .+= f_buffer             # du = f_buffer - K_buffer * u
    return nothing 
end

rhs_ode = ODEFunction(heat_equation_ode!, mass_matrix = M)

tspan = (0.0, 20.0)

params = ProblemParameters(K, M, f, dh, cellvalues, ch, K_buffer, f_buffer)
prob = ODEProblem(rhs_ode, u0, tspan, params)
sol = DifferentialEquations.solve(prob, Rodas5(autodiff=false),
            reltol = 1e-3, abstol = 1e-3, saveat=0.1);
```
